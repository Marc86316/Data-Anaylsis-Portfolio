## 💾 The data

The data comes from a survey hosted by an HR consultancy, available in `'salaries.csv'`.

#### Each row represents a single employee's salary record for a given year:
- **`work_year`** - The year the salary was paid.  
- **`experience_level`** - Employee experience level:  
  - **`EN`**: Entry-level / Junior  
  - **`MI`**: Mid-level / Intermediate  
  - **`SE`**: Senior / Expert  
  - **`EX`**: Executive / Director  
- **`employment_type`** - Employment type:  
  - **`PT`**: Part-time  
  - **`FT`**: Full-time  
  - **`CT`**: Contract  
  - **`FL`**: Freelance  
- **`job_title`** - The job title during the year.  
- **`salary`** - Gross salary paid (in local currency).  
- **`salary_currency`** - Salary currency (ISO 4217 code).  
- **`salary_in_usd`** - Salary converted to USD using average yearly FX rate.  
- **`employee_residence`** - Employee's primary country of residence (ISO 3166 code).  
- **`remote_ratio`** - Percentage of remote work:  
  - **`0`**: No remote work (<20%)  
  - **`50`**: Hybrid (50%)  
  - **`100`**: Fully remote (>80%)  
- **`company_location`** - Employer's main office location (ISO 3166 code).  
- **`company_size`** - Company size:  
  - **`S`**: Small (<50 employees)  
  - **`M`**: Medium (50–250 employees)  
  - **`L`**: Large (>250 employees)

## Data loading

Load the "salaries.csv" file into a pandas DataFrame.

```python
import pandas as pd

df = pd.read_csv('salaries.csv')
display(df.head())
```

## Data exploration

### Subtask:
Explore the dataset to understand its basic characteristics, specifically the number of records and the range of years covered.


**Reasoning**:
I need to determine the number of records and the range of years covered in the dataset.



```python
# Determine the number of records
num_records = len(df)

# Find the minimum and maximum years
min_year = df['work_year'].min()
max_year = df['work_year'].max()

# Print the results (for verification)
print(f"Number of records: {num_records}")
print(f"Range of years: {min_year} to {max_year}")
```

## Data analysis

### Subtask:
Calculate the average salary for Data Scientists and Data Engineers, and determine which role has a higher average salary.  Also, count the number of full-time employees based in the US who work 100% remotely.


**Reasoning**:
Calculate the average salaries for Data Scientists and Data Engineers, identify the higher-paying role, and count the number of full-time remote employees in the US.



```python
# Calculate average salaries
data_scientists = df[df['job_title'] == 'Data Scientist']
data_engineers = df[df['job_title'] == 'Data Engineer']

avg_data_scientist_salary = data_scientists['salary_in_usd'].mean()
avg_data_engineer_salary = data_engineers['salary_in_usd'].mean()

# Compare averages
if avg_data_scientist_salary > avg_data_engineer_salary:
    higher_paying_role = 'Data Scientist'
else:
    higher_paying_role = 'Data Engineer'

# Count remote workers
remote_workers = df[(df['employment_type'] == 'FT') &
                    (df['company_location'] == 'US') &
                    (df['remote_ratio'] == 100)]
num_remote_us_employees = len(remote_workers)

print(f"Average Data Scientist Salary: ${avg_data_scientist_salary:.2f}")
print(f"Average Data Engineer Salary: ${avg_data_engineer_salary:.2f}")
print(f"Higher Paying Role: {higher_paying_role}")
print(f"Number of remote US employees: {num_remote_us_employees}")
```
